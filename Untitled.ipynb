# IMPORTANT POINTS


## Features Explanation

* **age**: How old the patient is (in years)

* **sex**: Gender of the patient

  * 1 = male
  * 0 = female

* **cp (Chest Pain Type)**: Type of chest pain the patient feels

  * 0 = no serious pain
  * 1 = mild pain
  * 2 = unusual pain
  * 3 = severe/unclear pain

* **trestbps**: Blood pressure when resting

* **chol**: Amount of cholesterol in blood

* **fbs (Fasting Blood Sugar)**:

  * 1 = sugar level is high
  * 0 = normal sugar level

* **restecg**: Heart test (ECG) result when resting

  * 0 = normal
  * 1 = small problem
  * 2 = serious heart signal change

* **thalach**: Maximum heart rate achieved during exercise

* **exang**: Chest pain during exercise

  * 1 = yes
  * 0 = no

* **oldpeak**: How much heart stress increases during exercise

* **slope**: Heart activity pattern during exercise

  * 0 = improving
  * 1 = normal
  * 2 = getting worse

* **ca**: Number of blocked blood vessels (0–3)

* **thal**: Heart condition type

  * 1 = normal heart
  * 2 = fixed problem
  * 3 = temporary problem


##  Target Variable

* **target**: Final result

  * 0 = No heart disease
  * 1 = Heart disease present

```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
x,y,test_size=0.2,random_state=101)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
pd.DataFrame(x_train, columns=x.columns).describe()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)*100
print("Logistic Regreesion ")
print('------'*20)
print("Predict Heart Disease")
print('------'*20)
print("Accuracy",round(accuracy,2),"%")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

# Important Insights


* Dataset is clean with no missing values and duplicates were removed.
* Dataset is almost balanced between heart disease and no heart disease cases.
* Chest pain type is the strongest predictor of heart disease.
* Maximum heart rate (thalach) is strongly related to heart disease.
* Exercise-induced angina is strongly associated with heart disease cases.
* Higher values of major vessels (ca) are linked with lower chances of heart disease.
* Males have slightly more heart disease cases compared to females in this dataset.
* Patients with heart disease have a slightly lower median age (~52) compared to non-disease patients (~57).
* Age shows only a weak relationship and is not a strong standalone predictor.
* Logistic Regression achieved ~83% accuracy after scaling.
